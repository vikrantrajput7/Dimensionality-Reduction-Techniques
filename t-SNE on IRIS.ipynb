```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
```

```python
iris=pd.read_csv("iris.csv")
iris=iris.drop("Id",axis=1)
specs=iris["Species"]
iris=iris.drop("Species",axis=1)
```

```python
iris.head()
```

```python
tsne=TSNE(n_components=2,n_iter=300,perplexity=40)
x2d=tsne.fit_transform(iris)
```

```python
new_iris=pd.DataFrame(x2d,columns=["Feature1","Feature2"])

new_iris["Species"]=specs
```

```python
sns.pairplot(hue="Species",data=new_iris)
plt.show()
```

```python
sns.scatterplot(data=new_iris,x="Feature1",y="Feature2",hue="Species")
plt.show()
```

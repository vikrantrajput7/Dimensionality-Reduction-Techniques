```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("mnist.csv")
l=data["label"]
d=data.drop("label",axis=1)
```

```python
idx=30
img_grid=d.iloc[idx].values.reshape(28,28)
```

```python
plt.imshow(img_grid,cmap='gray')
print(l[idx])
plt.show()
```

# Applying PCA on MNIST Dataset

```python
from sklearn.decomposition import PCA
pca=PCA()
x2d=pca.fit(d)
```

```python
var_r=x2d.explained_variance_ratio_

plt.plot(var_r.cumsum())
plt.grid()
plt.show()
```

```python
x2d=PCA(n_components=2).fit_transform(d)

x2dt=pd.DataFrame(x2d,columns=["f1","f2"])

x2dt["Species"]=l

sns.scatterplot(data=x2dt,hue="Species",x="f1",y="f2")
plt.show()
```
